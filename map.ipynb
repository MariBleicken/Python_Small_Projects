```python
import folium
from folium import plugins
from folium.plugins import Geocoder
```

```python
sete_maravilhas_nomes = ["Grande Piramide de Gize", "Jardins Suspensos da Babilonia", "Templo de Artemis",
                         "Estatua de Zeus em Olipia", "Mausoleu de Halicarnasso","Colosso de Rodes","Farol de Alexandria"] 

sete_maravilhas_coordenadas = [[29.97796504524836, 31.13429790685845],
                              [30.738300,29.742780 ],
                              [32.542101, 44.421696],
                              [37.9499014743722, 27.364542836901638],
                              [37.63872560522192, 21.63180193336245], 
                              [37.038572207939694, 27.42544674102682],
                              [36.451066, 28.225833]]
```

```python
#Mapa

#Titulo
titulo = "Sete Maravilhas do Mundo Antigo"

titulo_html = '''
             <h1 align="center" style="font-size:30px"><b>{}</b></h1>
             '''.format(titulo)

sete_maravilhas_mapa = folium.Map(location = [33.9476, 27.6038], zoom_start = 5, control_scale=True, 
                                  tile = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Street_Map/MapServer/tile/{z}/{y}/{x}',
                                  attr = 'Tiles &copy; Esri &mdash; Source: Esri, DeLorme, NAVTEQ, USGS, Intermap, iPC, NRCAN, Esri Japan, METI, Esri China (Hong Kong), Esri (Thailand), TomTom, 2012',
                                  name = 'Esri.WorldStreetMap' )

sete_maravilhas_mapa.get_root().html.add_child(folium.Element(titulo_html))

#Adcionar caixa de busca

Geocoder(collapsed = True, position = "topleft").add_to(sete_maravilhas_mapa)

#Adcionar marcadores com as sete maravilhas
for i in range(len(sete_maravilhas_coordenadas)):
    coordenada = sete_maravilhas_coordenadas[i]
    maravilha = sete_maravilhas_nomes[i]
    folium.Marker(coordenada, popup=maravilha, icon = folium.Icon(icon = "glyphicon glyphicon-camera", 
                                                                  color = "black", 
                                                                  icon_color = "white", 
                                                                  prefix = "glyphicon")).add_to(sete_maravilhas_mapa)
    
#Visualizar coordenadas    
    sete_maravilhas_mapa.add_child(folium.LatLngPopup())
    
#Salvar mapa    
   
sete_maravilhas_mapa.save("sete_maravilhas.html")

sete_maravilhas_mapa
```
